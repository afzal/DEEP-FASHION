```python
import numpy as np

def step(z):
    return 1 if z >= Threshold else 0

def neuron(x, w):
    x = np.array(x)
    w = np.array(w)

    z = np.dot(w, x)   # sum of (wi * xi)
    y = step(z)        # activation (step)
    print("z =", z)
    return y

# Example:
x = [0.1, 0.5, 0.2]
w = [0.9, 0.1, 0.6]
Threshold = 0.4

# x = [1, 0.5, -2]
# w = [0.9, 1.0, -0.7]


print("Output y =", neuron(x, w))
```

```python
import numpy as np

def perceptron(x, w, b):
    z = np.dot(w, x) + b
    return 1 if z >= 0 else 0

# Example:
x = [0.1, 0.5, 0.2]
w = [0.4, 0.3, 0.6]

b = 0.5


# Example:
# x = [0.5, 1.2, -0.3]
# w = [1.0, -2.0, 0.8]
# b = 0.1

print(perceptron(x, w, b))
```
